```python
# Assistant
import pandas as pd
import matplotlib as mpl
mpl.rcParams['font.family'] = 'sans-serif'
mpl.rcParams['font.sans-serif'] = ['SimHei']

# 读取Excel文件中的豆瓣电影Top250数据表
df = pd.read_csv('豆瓣电影Top250.csv')

# 清洗上映年份列：移除所有非数字字符并转换为整数类型
df['上映年份'] = df['上映年份'].str.replace(r'\D', '', regex=True).astype(int)

# 清洗评分人数列：提取数字部分并转换为整数类型
df['评分人数'] = df['评分人数'].str.extract(r'(\d+)').astype(int)

# 生成数据概述信息，包括电影数量、年份范围和评分范围
overview = f"包含{len(df)}部电影，时间跨度{df['上映年份'].min()}-{df['上映年份'].max()}，评分范围{df['评分'].min()}-{df['评分'].max()}分"

# 定义并计算三个重要指标的详细分析数据
important_fields = {
    "评分": {
        "描述": "电影质量核心指标",
        "均值": df['评分'].mean(),
        "极差": df['评分'].max() - df['评分'].min(),
        "高分电影数量": len(df[df['评分'] >= 9.5])  # 统计评分9.5分以上的电影数量
    },
    "评分人数": {
        "描述": "电影热度指标",
        "最大值": df['评分人数'].max(),
        "最小值": df['评分人数'].min(),
        "中位数": df['评分人数'].median()  # 计算评分人数的中位数，反映典型热度水平
    },
    "上映年份": {
        "描述": "时间分布指标",
        "最新年份": df['上映年份'].max(),
        "最多年份": df['上映年份'].mode().values[0],  # 找出出现频率最高的年份
        "近十年电影": len(df[df['上映年份'] >= 2013])  # 统计2013年及以后的电影数量
    }
}

# 打印数据概述信息
print("1. 数据概述:", overview)
print("\n2. 重要指标分析:")
# 遍历每个指标并打印其详细信息
for field, info in important_fields.items():
    print(f"\n字段【{field}】- {info['描述']}")
    for k,v in info.items():
        if k != '描述':
            print(f"  {k}: {v}")

# 使用matplotlib创建可视化图表
import matplotlib.pyplot as plt

# 创建一个包含两个子图的图表
plt.figure(figsize=(12, 5))

# 第一个子图：电影评分的直方图
plt.subplot(1,2,1)
df['评分'].hist(bins=10, edgecolor='black')  # 创建10个区间的评分直方图
plt.title('电影评分分布')
plt.xlabel('评分')
plt.ylabel('数量')

# 第二个子图：电影上映年份的直方图
plt.subplot(1,2,2)
df['上映年份'].hist(bins=range(1950,2025,5), edgecolor='black')  # 以5年为间隔创建年份直方图
plt.title('电影年份分布')
plt.xlabel('上映年份')
plt.ylabel('数量')

# 调整布局并显示图表
plt.tight_layout()
plt.show()
```
